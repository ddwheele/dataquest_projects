# Visualizing Earnings Based on College Majors

This project uses [data](https://github.com/fivethirtyeight/data/tree/master/college-majors) published by FiveThirtyEight. It was collected by American Community Survey.

First let's look at the data.

```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

recent_grads = pd.read_csv('recent-grads.csv')
recent_grads.iloc[0]
```

```python
recent_grads.head()
recent_grads.size
```

```python
recent_grads.tail()
```

```python
recent_grads.describe()
```

```python
raw_data_count = recent_grads.shape[0]
```

```python
recent_grads = recent_grads.dropna()
cleaned_data_count = recent_grads.shape[0]
print(cleaned_data_count)
```

This is still too many majors to draw meaningful data from. Suppose we limited it to the most popular majors, I wonder if we would see stronger trends.

```python
popular = recent_grads[recent_grads['Sample_size'] > 250]
print(popular.shape[0])
```

```python
# Popularity of Major vs. Median Earnings
recent_grads.plot(x='Sample_size', y='Median', kind='scatter')
popular.plot(x='Sample_size', y='Median', kind='scatter')

# Unpopular majors can earn lots of money, or not, or anything in between.
```

```python
# Popularity of Major vs. Unemployment Rate
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter')
popular.plot(x='Sample_size', y='Unemployment_rate', kind='scatter')

# Unpopular majors can have any number of unemployed. The more unpopular, the greater the variation.
```

```python
# Graduates Employed Full Time vs. Median Earnings
recent_grads.plot(x='Full_time', y='Median', kind='scatter')
popular.plot(x='Full_time', y='Median', kind='scatter')

# Again, the fewer people employed, the greater the variation in earnings.
# Data do seem to indicate that very popular fields will earn less money.
```

```python
# Percent Women vs. Unemployment Rate
recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter')
popular.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter')

# The percent of women seems unrelated to the unemployment rate
```

```python
# Number of Men vs. Median Earnings
recent_grads.plot(x='Men', y='Median', kind='scatter')
popular.plot(x='Men', y='Median', kind='scatter')

# Once again, the fewer the numbers the higher the variation.
# It does look like if there are lots of men, earnings are lower.
```

```python
# Number of Women vs. Median Earnings
recent_grads.plot(x='Women', y='Median', kind='scatter')
popular.plot(x='Women', y='Median', kind='scatter')

# If lots of women, earnings are lower.
```

Hmm, looking at the data, it seems that restricting to popular majors did not bring out any significant trends.

```python
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]

fig = plt.figure(figsize=(5,12))
for r in range(4,8):
    ax = fig.add_subplot(4,1,r-3)
    ax = recent_grads[cols[r]].plot(kind='hist', rot=40)
```
